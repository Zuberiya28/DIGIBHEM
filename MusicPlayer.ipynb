```python
from pygame import mixer
from tkinter import *
from tkinter import filedialog

class MusicPlayer:
    def __init__(self, root):
        self.root = root
        self.root.title('Music Player')
        self.root.geometry("500x600")

        mixer.init()

        self.songs_list = Listbox(self.root, bg="purple", fg="white", font=('arial', 12), height=20, width=50, selectbackground="pink", selectforeground="blue")
        self.songs_list.grid(row=0, column=0, columnspan=4, pady=10)

        self.play_button = Button(self.root, text="Play", width=7, command=self.play_song)
        self.play_button.grid(row=1, column=0, padx=5)

        self.pause_button = Button(self.root, text="Pause", width=7, command=self.pause_song)
        self.pause_button.grid(row=1, column=1, padx=5)

        self.stop_button = Button(self.root, text="Stop", width=7, command=self.stop_song)
        self.stop_button.grid(row=1, column=2, padx=5)

        self.resume_button = Button(self.root, text="Resume", width=7, command=self.resume_song)
        self.resume_button.grid(row=1, column=3, padx=5)

        self.add_song_button = Button(self.root, text="Add Songs", width=10, command=self.add_songs)
        self.add_song_button.grid(row=2, column=1, pady=5)

        self.delete_song_button = Button(self.root, text="Delete Song", width=10, command=self.delete_song)
        self.delete_song_button.grid(row=2, column=2, pady=5)

    def add_songs(self):
        songs = filedialog.askopenfilenames(initialdir="Music/", title="Choose songs", filetypes=(("mp3 Files", "*.mp3"),))
        for song in songs:
            song=song.replace("C:/Users/SIFY/Desktop/musicplayer/Songs","")
            self.songs_list.insert(END, song)

    def delete_song(self):
        selected_song = self.songs_list.curselection()
        if selected_song:
            self.songs_list.delete(selected_song[0])

    def play_song(self):
        selected_song = self.songs_list.curselection()
        if selected_song:
            song_path = self.songs_list.get(selected_song)
            song_path=f'C:/Users/SIFY/Desktop/musicplayer/Songs/{song_path}'
            mixer.music.load(song_path)
            mixer.music.play()

    def pause_song(self):
        mixer.music.pause()

    def stop_song(self):
        mixer.music.stop()

    def resume_song(self):
        mixer.music.unpause()

if __name__ == "__main__":
    root = Tk()
    music_player = MusicPlayer(root)
    root.mainloop()
```
